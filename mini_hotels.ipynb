## Цель исследования:

Цель данного проекта — заключается в определении средней стоимости строительства мини-отелей/гостиниц в Российской Федерации. В рамках исследования будут решены следующие задачи: анализ влияния региона на стоимость строительства, оценка влияния стоимости строительных материалов и услуг на общий бюджет, а также выявление ключевых факторов, определяющих вариативность цен в различных частях страны.

###Актуальность исследования:

**Тренды и прогнозы на 2024 год**  

- Российский туризм ожидает рост спроса на внутренние направления и более раннее бронирование поездок.  
- Популярны направления в Юго-Восточной Азии, особенно Таиланд и Вьетнам, а также Турция и Египет для летнего отдыха​ (B2B.Ostrovok)​​ (vc.ru)​.  
- Выездной туризм увеличивается, с акцентом на поездки в СНГ и Азию​ (IZ.RU)​.    
- Внутренний туризм также набирает популярность, особенно поездки в Краснодарский край, Санкт-Петербург, Москву и Дагестан​ (vc.ru)​.  
<br>  
Спрос на курорты и конкретные направления остается высоким, особенно на популярные направления как в России, так и за рубежом​ (IZ.RU)​​ (vc.ru)​.

В условиях растущего спроса на внутренний туризм и популярности курортных направлений, знание стоимости строительства мини-отелей в РФ имеет важное значение для:

**Инвесторов и предпринимателей:** Поможет оценить экономическую целесообразность проектов.  
**Разработчиков туристической инфраструктуры:** Обеспечит понимание затрат на создание конкурентоспособного продукта.  
**Местных властей:** Содействует планированию и развитию региональной туристической инфраструктуры.  
<br>  
Знание стоимости строительства мини-отелей позволит учесть данный фактор в стратегическом развитии туристического сектора и удовлетворять растущий спрос на внутренние поездки.

***Примечание: данное исследование не включает в себя исследование стоимости земельных участков, стоимости рабочей силы, стоимости сопутствующих материалов, расходов на оснащение отлей и тп.***

На основании выведенной информации наблюдается:  
- в колонках "square_2", "discount", "square_1", "url", "levels" и "rooms" значительная доля информации отсутствует, восстановление всех значений данных колонок не имеет смысла, так как данная информация в целях текущего анализа использоваться не будет. Данные колонки можно удалить, но оставим их без изменения так как они нам не помешают;  
- колонка "space" имеет тип object, необходимо поменять тип;   
- колонки "levels" и "rooms" имеют тип float, можно оставить без изменений.


```python
df.space.unique()
```

```python
df.space = df.space.str.replace('м2', ' ')
df.space = df.space.str.replace('311   + 88  ', '399')
df.space = df.space.str.replace('1.364.00', '1364')
df.space = df.space.str.replace('1.291.86', '1291.86')
```

```python
df.space.unique()
```

```python
df['space'] = df['space'].astype('float')
```

Для дальнейшего анализа целесообразно добавить колонку с стоимостью за кв.м.

```python
df.insert(4, 'price_m_sq', df['price'] / df['space'])
df.head()
```

Представленный датафрейм имеет 312 строк и 13 колонок (12 изначальных и одна колонка была добавлена), колонки содержат количественные и категориальные переменные.

Вторая гипотеза утверждает, что средняя цена на мини-отели не зависит от материала строительства, предполагая, что выбор строительного материала не оказывает значительного влияния на общую стоимость.

По столбцу цена наблюдаются значительные выбросы, удалим их.

```python
# Функция для подсчёта границ с учетом 1.5 размаха
def quartile_range(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq
# Столбцы с выбросами
list_features = ['price_m_sq']
for col in list_features:
    df[col] = quartile_range(df, col)
```
